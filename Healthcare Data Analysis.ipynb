```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate a synthetic healthcare dataset
np.random.seed(42)

# Number of samples
n_samples = 500

# Generate random patient data
patient_data = pd.DataFrame({
    'PatientID': range(1, n_samples + 1),
    'Gender': np.random.choice(['Male', 'Female'], size=n_samples),
    'Age': np.random.randint(18, 80, size=n_samples),
    'Disease': np.random.choice(['A', 'B', 'C', 'D'], size=n_samples),
    'TreatmentDuration': np.random.randint(1, 30, size=n_samples),
    'Outcome': np.random.choice(['Recovered', 'Not Recovered'], size=n_samples)
})

# Display the first few rows of the synthetic dataset
print(patient_data.head())

# Summary statistics
summary_stats = patient_data.describe()
print(summary_stats)

# Explore patient demographics
sns.countplot(x='Gender', data=patient_data)
plt.title('Distribution of Gender in the Synthetic Dataset')
plt.show()

# Analyze disease prevalence
top_diseases = patient_data['Disease'].value_counts().head(5)
top_diseases.plot(kind='bar')
plt.title('Top 5 Most Prevalent Diseases in the Synthetic Dataset')
plt.xlabel('Disease')
plt.ylabel('Number of Cases')
plt.show()

# Visualize treatment outcomes
plt.figure(figsize=(10, 6))
sns.boxplot(x='Outcome', y='TreatmentDuration', data=patient_data)
plt.title('Distribution of Treatment Duration by Outcome in the Synthetic Dataset')
plt.xlabel('Outcome')
plt.ylabel('Treatment Duration (days)')
plt.show()

# Correlation heatmap
correlation_matrix = patient_data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix for Numerical Variables in the Synthetic Dataset')
plt.show()
```
